```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
j15 = pd.read_csv('./data/comunio_J15.csv')
j16 = pd.read_csv('./data/comunio_J16.csv')
j17 = pd.read_csv('./data/comunio_J17.csv')
j18 = pd.read_csv('./data/comunio_J18.csv')

```

```python
j18 = j18.drop(['Last_5_games_points'], axis=1)

```

```python

j18['J_11'] = j15['J_11']
j18['J_12'] = j15['J_12']
j18['J_13'] = j15['J_13']

j18['Value_J_15'] = j15['Value']
j18['Value_J_16'] = j16['Value']
j18['Value_J_17'] = j17['Value']
```

```python
'''gk = j17.loc[j17['Position']=='PT'].sort_values('Total_Points', ascending=False)[1:6]
df = j17.loc[j17['Position']=='DF'].sort_values('Total_Points', ascending=False)[0:11]
md = j17.loc[j17['Position']=='MD'].sort_values('Total_Points', ascending=False)[:15]
dl = j17.loc[j17['Position']=='DL'].sort_values('Total_Points', ascending=False)[:15]'''

gk = j18.loc[j18['Position']=='PT'].sort_values('Total_Points', ascending=False)[0:5]
df = j18.loc[j18['Position']=='DF'].sort_values('Total_Points', ascending=False)[0:10]
md = j18.loc[j18['Position']=='MD'].sort_values('Total_Points', ascending=False)[:10]
dl = j18.loc[j18['Position']=='DL'].sort_values('Total_Points', ascending=False)[:10]
```

```python
top_players = pd.DataFrame(gk)
```

```python
top_players = top_players.append(df)
```

```python
top_players = top_players.append(md)
top_players = top_players.append(dl)
```

```python
top_players.info()
```

```python
top_players.columns
```

```python
new_columns = ['Team_id', 'Team', 'Player', 'Position', 'Matches_Played', 'On_start_%',
       'Goals_/_Goals_Received', 'Assists', 'Total_Points', 'Points_Average',
       'J_11', 'J_12','J_13', 'J_14', 'J_15', 'J_16', 'J_17','J_18', 'Avg_last_5_games',
        'Value_J_15', 'Value_J_16' ,'Value_J_17','Value']
```

```python
top_players = top_players.reindex(columns=new_columns)
top_players.info()
```

```python
top_players.J_11 = top_players.J_11.astype(int)
top_players.J_12 = top_players.J_12.astype(int)
top_players.J_13 = top_players.J_13.astype(int)
top_players.Value_J_15 = top_players.Value_J_15.astype(int)
top_players.Value_J_16 = top_players.Value_J_16.astype(int)
top_players.Value_J_17 = top_players.Value_J_17.astype(int)
```

```python
#top_players = top_players.loc[top_players['Matches_Played']!=0]
```

```python
top_players = top_players.rename(columns ={'J_11':'2021-10-26', 'J_12':'2021-10-30', 'J_13':'2021-11-05', 'J_14':'2021-11-19', 
                                  'J_15': '2021-11-26', 'J_16':'2021-12-03', 'J_17':'2021-12-10','J_18':'2021-12-17', 'Value_J_15': '2021-11-26(V)', 
                                           'Value_J_16':'2021-12-03(V)', 'Value_J_17':'2021-12-10(V)', 'Value':'2021-12-17(V)'})
```

```python
top_players['Price_per_Total_Points'] = top_players['2021-12-17(V)']/top_players.Total_Points
```

```python
ppp = []
for value in top_players.Price_per_Total_Points:

    value = '{0:.2f}'.format(value)
    
    ppp.append(value)
```

```python
top_players.Price_per_Total_Points = ppp
```

```python
Price_per_Points_Last_5_J = top_players['2021-12-17(V)']/ (top_players['2021-12-17']+top_players['2021-12-10']+ top_players['2021-12-03']+top_players['2021-11-26']+top_players['2021-11-19'])

pppl5j = []

for value in Price_per_Points_Last_5_J:
    value = '{0:.2f}'.format(value)
    pppl5j.append(value)

top_players['Price_per_Point_Last_5_J']= pppl5j
```
