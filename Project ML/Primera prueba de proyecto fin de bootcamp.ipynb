```python
import pandas as pd
from fuzzywuzzy import process, fuzz
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
lr = LogisticRegression()
rfr = RandomForestRegressor()
dt = DecisionTreeRegressor()
xgbr = XGBRegressor()
svr = SVR()
gb = GradientBoostingRegressor()

models = (lr, rfr, dt, xgbr, svr, gb)
```

```python
comunio = pd.read_csv('../data/pruebas/only_comunio_stats_J27.csv')
fbref = pd.read_csv('../data/pruebas/only_fbref_stats_J27.csv')
```

```python
comunio
```

```python
comunio['Total_Points'] = comunio.Total_Points- (comunio.J_26+comunio.J_27)
comunio['Points_Average'] = comunio.Total_Points/25
comunio
```

```python
comunio = comunio.drop(['Team_id' ,'Last_5_games_points', 'Avg_last_5_games', 'J_27'], axis=1)
```

```python
comunio
```

```python
X = comunio.drop('J_26',axis=1)._get_numeric_data()
y = comunio.J_26
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2)
for model in models:
    name = str(model)[:13]
    if name == 'LogisticRegre':
        scalar = StandardScaler().fit(X_train)
        model.fit(scalar.transform(X_train),y_train)
        model.score(scalar.transform(X_train), y_train)
        pred = model.predict(scalar.transform(X_test))
        mse = mean_squared_error(y_test, pred, squared=False)
    else:    
        model.fit(X_train,y_train)
        model.score(X_train, y_train)
        pred = model.predict(X_test)
        mse = mean_squared_error(y_test, pred, squared=False)
    
    print (f'mse del modelo {name} = {mse}')
```

```python
df_1= pd.DataFrame()
df_1['Player'] = comunio.Player
df_1['Target'] = comunio.J_26
df_1['Pred_svr'] = (svr.predict(X))
df_1['Pred_rfr'] = rfr.predict(X)
df_1['Pred_xgbr'] = xgbr.predict(X)
df_1['Pred_lr'] = lr.predict(X)
df_1['Pred_dt'] = dt.predict(X)
df_1['Pred_gb'] = gb.predict(X)
df_1 = round(df_1)
df_1
```

```python
mean_squared_error(y, df_1.Pred_xgbr, squared=False)
```

```python
df_1['Test'] = df_1.Target - df_1.Pred_xgbr
df_1
```

```python
def once_ideal_xbgr(data):
    gk = data.loc[data['Position']=='PT'].sort_values('Predict_xbgr', ascending=False)[:1]
    df = data.loc[data['Position']=='DF'].sort_values('Predict_xbgr', ascending=False)[:3]
    md = data.loc[data['Position']=='MD'].sort_values('Predict_xbgr', ascending=False)[:4]
    dl = data.loc[data['Position']=='DL'].sort_values('Predict_xbgr', ascending=False)[:3]
    positions = [gk, df, md, dl]
    squad = pd.DataFrame()
    for pos in positions:
        squad = squad.append(pd.DataFrame(pos))
    
    return squad
```
