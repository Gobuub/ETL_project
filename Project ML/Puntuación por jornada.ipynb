```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
pts = [59,45,78,67,52,50,48,28,50,45,40,72,44,66,63,64,71,54,60,60,30,52,95,45,89,67,55,71,81,71,50]
jornadas = [i for i in range(1,32)]

dtc = dict(zip(jornadas,pts))
```

```python
df = pd.DataFrame(pts, columns=['Pts'], index=jornadas)
```

```python
values = df.values
nothing, etl, ml = values[0:15, :], values[14:27,:], values[26:, :]
```

```python
index = df.index.values
plt.figure(figsize=(15,4))
plt.plot(index[0:15],nothing,c='green')
plt.plot(index[14:27],etl,c='blue',alpha=0.7)
plt.plot(index[26:],ml,c='orange',alpha=0.7)
plt.legend(['Nothing','ETL', 'ML'])
plt.axvline(15, c="r")
plt.axvline(27, c="r")
plt.grid()
```
