```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python

j20 = pd.read_csv('./data/pruebas/comunio_J20.csv')

```

```python
j20 = j20.drop(['Last_5_games_points'], axis=1)

```

```python

gk = j20.loc[j20['Position']=='PT'].sort_values('Total_Points', ascending=False)[0:5]
df = j20.loc[j20['Position']=='DF'].sort_values('Total_Points', ascending=False)[0:10]
md = j20.loc[j20['Position']=='MD'].sort_values('Total_Points', ascending=False)[:10]
dl = j20.loc[j20['Position']=='DL'].sort_values('Total_Points', ascending=False)[:10]
```

```python
top_players = pd.DataFrame(gk)
```

```python
top_players = top_players.append(df)
```

```python
top_players = top_players.append(md)
top_players = top_players.append(dl)
```

```python
top_players.info()
```

```python
top_players = top_players.rename(columns ={ 'J_16':'2021-12-03', 'J_17':'2021-12-10','J_18':'2021-12-17', 'J_19':'2021-12-21', 'J_20': '2022-01-01','Value':'2022-01-01(V)'})
```

```python
top_players['Price_per_Total_Points'] = top_players['2022-01-01(V)']/top_players.Total_Points
```

```python
ppp = []
for value in top_players.Price_per_Total_Points:

    value = '{0:.2f}'.format(value)
    
    ppp.append(value)
```

```python
top_players.Price_per_Total_Points = ppp
```

```python
Price_per_Points_Last_5_J = top_players['2022-01-01(V)']/ (top_players['2021-12-21']+top_players['2021-12-17']+ top_players['2021-12-10']+top_players['2021-12-03']+top_players['2022-01-01'])

pppl5j = []

for value in Price_per_Points_Last_5_J:
    value = '{0:.2f}'.format(value)
    pppl5j.append(value)

top_players['Price_per_Point_Last_5_J']= pppl5j
```
