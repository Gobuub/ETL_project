```python
def comunio_stats(team_lst, jornada):
    
    df_teams = pd.DataFrame()
    
    for team in team_lst:
        print(team[0])
        soup = bs(req.get(team[2]).text , 'html.parser')
        soup2 = bs(req.get(team[4]).text , 'html.parser')
        soup3 = bs(req.get(team[5]).text , 'html.parser')

        jugadores = team[3] # necesario porque no todas las plantillas tienes los mismos jugadores, entra por parámetro

        team_name = team[0]
        team_id = team[1]

        equipo = team_name #* jugadores

        name = soup.find_all('strong')
        name = [name[i].text for i in range(22, (22 + jugadores * 2), 2)]

        pos_gk = soup.find_all('span', class_='label-posicion label-success')
        gks = [pos_gk[i].text.strip() for i in range(len(pos_gk))]

        pos_df = soup.find_all('span', class_='label-posicion label-info')
        dfs = [pos_df[i].text.strip() for i in range(len(pos_df))]

        pos_md = soup.find_all('span', class_='label-posicion label-warning')
        mds = [pos_md[i].text.strip() for i in range(len(pos_md))]

        pos_fw = soup.find_all('span', class_='label-posicion label-danger')
        fws = [pos_fw[i].text.strip() for i in range(len(pos_fw))]

        all_pos = gks + dfs + mds + fws

        pts = soup.find_all('span', class_='texto_pequenio2 label-posicion label-primary')

        pt = [pts[i].text for i in range(len(pts))]

        values = soup.find_all('small')
        value = [values[i].text.split('€')[0] for i in range(0,len(values),3)]

        pt_avg = [values[i].text.split()[-1] for i in range(1,len(values),3)]

        on_start_per = [values[i].text.split('%')[0] for i in range(2,len(values),3)]

        puntos = soup2.find_all('div', class_='streak')
        pts_last_five_jouyneys = [puntos[i].text.split() for i in range(len(puntos))] 
        
        jback5 = []
        jback4 = []
        jback3 = []
        jback2 = []
        jback1 = []
        avg_last_5 = []
        for i, ptos in enumerate(pts_last_five_jouyneys):
            avg = 0/5
            if pts_last_five_jouyneys[i][0] == '-':
                jback5.append(0)
            else:
                jback5.append(int(pts_last_five_jouyneys[i][0]))
                avg += int(pts_last_five_jouyneys[i][0])
            if pts_last_five_jouyneys[i][1] == '-':
                jback4.append(0)
            else:
                jback4.append(int(pts_last_five_jouyneys[i][1]))
                avg += int(pts_last_five_jouyneys[i][1])
            if pts_last_five_jouyneys[i][2] == '-':
                jback3.append(0)
            else:
                jback3.append(int(pts_last_five_jouyneys[i][2]))
                avg += int(pts_last_five_jouyneys[i][2])
            if pts_last_five_jouyneys[i][3] == '-':
                jback2.append(0)
            else:
                jback2.append(int(pts_last_five_jouyneys[i][3]))
                avg += int(pts_last_five_jouyneys[i][3])
            if pts_last_five_jouyneys[i][4] == '-':
                jback1.append(0)
            else:
                jback1.append(int(pts_last_five_jouyneys[i][4]))
                avg += int(pts_last_five_jouyneys[i][4])

            avg_last_5.append(avg/5)        


        jornada = jornada

        team = {#'Team_id': team_id, #entra como parametro en la función
                'Team': equipo, # cambiar por el parametro de entrada de entrada team
                'Player': name,
                'Position': all_pos,
                'Total Points': pt,
                'Points Average': pt_avg,
                'Value': value,
                'On start %': on_start_per,
                'Last 5 matchs points': pts_last_five_jouyneys,
                f'J{jornada - 4}': jback5,
                f'J{jornada - 3}': jback4,
                f'J{jornada - 2}': jback3,
                f'J{jornada - 1}': jback2,
                f'J{jornada}': jback1,
                'Avg ultimas 5 J': avg_last_5}
        
        
        '''
        team = {#'Team_id': team_id, #entra como parametro en la función
                'Team': equipo, # cambiar por el parametro de entrada de entrada team
                'Player': name,
                'Position': all_pos,
                'Total Points': pt,
                'Points Average': pt_avg,
                'Value': value,
                'On start %': on_start_per,
               'Last 5 matchs points': pts_last_five_jouyneys }

        #print(equipo)
        #print(len(name))
        #print(len(all_pos))
        #print(len(pt))
        #print(team)
        '''
        df = pd.DataFrame(team)
        #df
        #df = pd.DataFrame(team)
        #df.to_csv(f'{team_name}.csv', index = False)
        #data.append(team)
        
        # Segunda parte del scrapping para hacer la tabla de estadisticas
        
        #espn_url = team[5]

        #soup3 = bs(req.get(team[5]).text , 'html.parser')

        team_name = soup3.find('span', class_='db fw-bold').text.strip()

        porteros = soup3.find('div', class_='Table__Scroller')
        filas_porteros = porteros.find_all('tr')
        stats_porteros = [[e.text for e in filas_porteros[i].find_all('td')] for i in range(1,len(filas_porteros))]

        jugadores_campo = soup3.find_all('div', class_='Table__Scroller')
        filas_jugadores_campo = jugadores_campo[1].find_all('tr')
        stats_jugadores = [[e.text for e in filas_jugadores_campo[i].find_all('td')] for i in range(1,len(filas_jugadores_campo))]

        pl_lst = []

        for i, player in enumerate(stats_porteros):
            name = ''.join([e for e in stats_porteros[i][0] if e.isalpha() or e == ' '])
            matchs = player[6]
            goles_encajados = player[9]
            asistencias = player[10]


            gk_dict = {'Player': name, 'Matchs': matchs, 'Goals / Goals Received': goles_encajados, 'Assistances' : asistencias}
            pl_lst.append(gk_dict)

        for i, player in enumerate(stats_jugadores):
            name = ''.join([e for e in stats_jugadores[i][0] if e.isalpha() or e == ' '])
            matchs = player[6]
            goles = player[8]
            asistencias = player[9]


            pl_dict = {'Player': name, 'Matchs': matchs, 'Goals / Goals Received': goles, 'Assistances' : asistencias}
            pl_lst.append(pl_dict)
        
        df_stats = pd.DataFrame(pl_lst)
        
        df[['team_matched', 'fuzz_score']]=df.Player.apply(lambda x:process.extractOne(x,
                                                                     df_stats.Player.tolist(),
                                                                     scorer=fuzz.partial_ratio)).apply(pd.Series)
        df=pd.merge(df, df_stats, left_on='team_matched', right_on='Player')
        df_1=df[df.fuzz_score>85]
        df_1 = df_1.drop(columns = ['team_matched', 'fuzz_score', 'Player_y'])
        df_1 = df_1.rename(columns = {'Player_x': 'Player'})
        order_col = [0,1,2,14,6,15,16,3,4,7,8,9,10,11,12,13,5]
        df_1 = df_1[df_1.columns[order_col]]
        df_teams = df_teams.append(df_1)
        df_1.to_csv(f'data/{team_name}_J{jornada}.csv', index=False)
    
    
    df_teams.to_json(f'data/comunio_J{jornada}.json', orient="table")
    df_teams.to_csv(f'data/comunio_J{jornada}.csv', index=False)
    
    return 'Finished'
```
